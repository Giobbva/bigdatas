```python
import pandas as pd
import pyarrow.parquet as pq
```

```python
# Leer archivo parquet
table = pq.read_table("0000.parquet")
```

```python
df = table.to_pandas()
print(df)
```

```python
print("Shape:", df.shape)
```

```python
pd.set_option("display.max_rows", None)      # Mostrar todas las filas
pd.set_option("display.max_columns", None)   # Mostrar todas las columnas
pd.set_option("display.width", None)         # No cortar por ancho

# Visualizar el dataframe completo
print(df)
```

```python
df.to_csv("mi_dataset.csv", index=False)

print("✅ Archivo CSV guardado como 'mi_dataset.csv'")
```

```python
print(df.info())
print("\nPrimeras 5 filas:\n", df.head())
```

```python
# Obtener la segunda columna (índice 1)
col2 = df.iloc[:, 1]

# Obtener los valores únicos
temas_unicos = col2.unique()

# Imprimir resultados
print("Temas únicos en la columna 2:")
for tema in temas_unicos:
    print("-", tema)

print(f"\nTotal de temas diferentes: {len(temas_unicos)}")
```
